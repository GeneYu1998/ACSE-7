#### Modifications to forward model

Given the observational data, we should first consider whether any modifications to the forward model will simplify our lives. 

The existing forward model already outputs at the observation locations, so we can use a (fairly) trivial observation operator. If we choose to increase the model resolution, this is a useful quantity to maintain (i.e. increase it by a whole number factor like double or triple rather than a fractional increase). The model also outputs on a timescale which matches the observations, so time-aware schemes like 4D-Var can assimilate on the timestep rather than needing to interpolate between them, and whole time level schemes like OI can work trivially. Again, if we need to increase the time resolution, we should definitely keep this property.

If we are using 4D-Var, one key modification that is worth making is to output the  model data on every timestep. Since the model is nonlinear, we will need it for the adjoint equation (or to generate an approximation via interpolation). It's also useful to store the validity times of the model data.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import inv

data = np.load('burgers_observations.npz')
t_obs = data['t_obs']
y = data['y']

# B is the (sparse) background error covariance matrix

def generate_B_matrix(z):
    B = np.empty([len(z), len(z)])
    
    for _ in range(len(z)):
        B[_, :] = 0.2**2*np.exp(-4*(z[_]-z)**2)*((z[_]-z)**2<=4)
    return B

nz = 61
z = np.linspace(-10, 10, nz)
dz = z[1]-z[0]
u = np.full(z.shape, 1.0)
nu = 0.4
dt = 0.05

B = generate_B_matrix(z)

## Use forward Euler for the (nonlinear) advection  and Crank-Nicolson for diffusion
def rhs(U, dt, t):
    out = np.empty(U.shape)
    out[0] = U[0]-0.5*dt/dz*((U[1]**2-U[-1]**2)/2.0-nu*(U[-1]-2*U[0]+U[1])/dz)
    out[1:-1] = U[1:-1]-0.5*dt/dz*((U[2:]**2-U[:-2]**2)/2.0-nu*(U[:-2]-2*U[1:-1]+U[2:])/dz)
    out[-1] = U[-1]-0.5*dt/dz*((U[0]**2-U[-2]**2)/2.0-nu*(U[-2]-2*U[-1]+U[0])/dz)
    
    return out+dt*10*np.sin(.375*np.pi*(t+0.5*dt))*np.exp(-100*z**2)

def lhs_matrix(u, dt):
    mat = np.eye(u.shape[0])
    mat[0,-1] -= 0.5*dt*nu/dz**2
    mat[0, 0] += dt*nu/dz**2
    mat[0, 1] -= 0.5*dt*nu/dz**2
    for _ in range(1, u.shape[0]-1):
        mat[_, _-1] -= 0.5*dt*nu/dz**2
        mat[_, _] += dt*nu/dz**2
        mat[_, _+1] -= 0.5*dt*nu/dz**2
    mat[-1,-2] -= 0.5*dt*nu/dz**2
    mat[-1, -1] += dt*nu/dz**2
    mat[-1, 0] -= 0.5*dt*nu/dz**2
    return mat



def burgers_equation(u0, dt, t_final, t_0=0.0):
    """ Burger's equation solver"""
    t = t_0

    u = u0
    out = [u0]
    times = [t]
    mat = lhs_matrix(u0, dt)
    matinv = inv(mat)
    while t<t_final:
        u = matinv.dot(rhs(u, dt, t))
        t += dt
        out.append(u)
        times.append(t)
    return out, times

data = burgers_equation(u, dt, 100)[0]
plt.plot(z, data[-1]);
```

#### Continous Adjoint 4D-Var

For this method we first generate the tangent linear and adjoint models for the equation, then build a numerical model to solve them. The tangent linear model can be written down by hand, but we can also use `sympy` to solve for the identity

$$ L(\delta u) = \left[\frac{d}{da} \mathcal{L}(u+a\delta u)\right]_{a=0}, $$

where $\mathcal{L}$ is a non-linear model and $L$ is our tangent linearization.

```python
import sympy

z = sympy.Symbol('z', real=True)
t = sympy.Symbol('t', real=True)
a = sympy.Symbol('a', real=True)
nu = sympy.Symbol(r'\nu', real=True)
u = sympy.Function('u')(z,t)
du = sympy.Function('du')(z,t)
f = sympy.Function('f')(z,t)

#Burgers' equation

L = sympy.Derivative(u, t) + sympy.Derivative(u**2, z)/2- nu*sympy.Derivative(u, z, z)-f

sympy.init_printing()

# The linear model
sympy.Equality(sympy.Derivative(L.subs(u, u+a*du),
                                a, evaluate=True).subs(a,0), 0)
```

Writing this in a more human readable order, the tangent linear equation is

$$ \frac{\partial}{\partial t} \delta u + u \frac{\partial}{\partial z} \delta u +\delta u \frac{\partial u}{\partial z} = \nu\frac{\partial ^2 }{\partial z^2}  \delta u,$$
or any of the variety of rearrangements of the same. 

Note that the forcing term **does not** appear in the tangent linear model for the perturbation since the forcing doesn't have any terms in $u$ in it. This is not that surprising, since the Taylor expansion is
$$ \mathcal{L}(u+a\delta u) = \mathcal{L}(u) + a L(\delta u) +\mathcal{O}(a^2) $$
and the forcing is already balanced by the background trajectory.

The adjoint model can be found be writing down
$$ \mathcal{J}_{ext} := \mathcal{J}+\int \int \lambda L(\delta u) dz\,dt = \mathcal{J} + \left[\int \lambda \delta u\, dz\right]_{t_0}^{t_f} - \int \int \delta u L^\dagger (\lambda) dz \, dt. $$

The third term on the right-hand side is derived by integration by parts of both the time derivative term and the $\partial/\partial z$ terms which are combined in a linear operator $L^\dagger(\lambda)$. The adjoint equation $L^\dagger(\lambda)=0$ is then obtained by requiring the variation in the functional $\mathcal{J}$ to vanish for arbitrary $\delta u$ (i.o.w. the derivative w.r.t $u$ is zero). Written out $L^\dagger(\lambda)=0$ produces the following PDE for $\lambda$:

$$ \frac{\partial \lambda}{\partial t} + \frac{\partial(u\lambda) }{\partial z} - \lambda\frac{\partial u }{\partial z}= -\nu\frac{\partial ^2 \lambda}{\partial z^2}  $$
or a rearragement of the same
$$ \frac{\partial \lambda}{\partial t} + u \frac{\partial\lambda }{\partial z} = -\nu\frac{\partial ^2 \lambda}{\partial z^2}  $$.

Integration by parts of the time derivative term also gives rise to the second term on the right-hand side, which involves a integral over $z$ at $t=t_0$ and $t=t_f$. The integral at $t=t_f$ vanishes if we use $\lambda=0$ at $t=t_f$ as the "initial" (or should we say final) condition for the backwards PDE. 

In other words in the gaps between observations, sensitivities are advected backwards with speed $-u$ and diffuse at the rate $\nu$.


Suitable code for this might look very similar an unforced version of the forward model

```python
## Let's discretize using a simple "Forward Euler"  to solve backwards in time.
## This is simple to code, but definitely *NOT* a good choice, due to its stability
## issues.

## In general the adjoint equation is likely to be at least as stable as the forward
## model, but the Burger's equation is known to be stiff, due to its tendency to form shocks.

z = np.linspace(-10, 10, nz)
dz = z[1]-z[0]
nu = 0.4

def adjoint_rhs(U, l, dt, *args):
    """ Calculate the update for a Forward Euler discretization of the continuous
    adjoint equation.
    
    Parameters
    ----------
    U : arraylike
        The Burgers state being linearized around
    l : arraylike
        The perturbation variable
    dt :  float
        The timestep
        
    Returns
    -------
    l_new : arraylike
        the new value of the perturbation variable.
    
    """    
    
    out = np.empty(U.shape)
    out[0] = l[0]+dt/dz*(U[0]*(l[1]-l[-1])/2.0
                             +nu*(l[-1]-2*l[0]+l[1])/dz)
    out[1:-1] = l[1:-1]+dt/dz*(U[1:-1]*(l[2:]-l[:-2])/2.0
                                   +nu*(l[:-2]-2*l[1:-1]+l[2:])/dz)
    out[-1] = l[-1]+dt/dz*(U[-1]*(l[0]-l[-2])/2.0
                               +nu*(U[-2]-2*U[-1]+U[0])/dz)
    
    return out

def continuous_adjoint_equation(u0, dt, t_final, y, t_obs, t_0=0.0):
    """ Solve the continuous adjoint equation for a 4d var model.
    
    Parameters
    ----------
    u0 : arraylike
        The starting condition
    dt :  float
        The timestep
    t_final: float
        The final time
    y : list
        a collection of observations
    t_obs: list
        the validity times of the observations
    t_0 : float
        the start time
        
    Returns
    -------
    l :
    The update valid at t_0
    """    
    u, times = burgers_equation(u0, dt, t_final, t_0)
    l = 0*u[-1]
    mat = lhs_matrix(l, dt)
    matinv = inv(mat)        
    
    for U, t in zip(u[::-1], times[::--1]):
            if np.isclose(t, t_obs[-1]):
                l -= H.T@Rinv@(y[-1]-H@U)
                t_obs = t_obs[:-1]
                y = y[:-1]
            l = adjoint_rhs(U, l, dt, t)
    return l

## Now define the standard 4d-var cost function and its gradient
z = np.linspace(-10, 10, nz)
B = generate_B_matrix(z)
Binv = inv(B)
Rinv = inv(R)

def J_4d_var(u, u_b, dt, y, t_obs, t_0):
    """Calculate 4D-Var cost function
    
    Parameters
    ----------
    
    u : arraylike
        test state
    u_b : arraylike
        background state
    dt : float
        timestep
    y : list
        Collection of observations
    t_obs: arraylike
        Validity times for the observations
    t_0 : float
        start time
    
    Returns
    -------
    J : float
        4D-var cost function value
    
    """
    U, times = burgers_equation(u, dt, t_obs[-1], t_0)
    J_0 = (u-u_b)@Binv@(u-u_b)/2.0
    J_obs = 0.0
    k = 0
    for Y, t in zip (y, t_obs):
        for _, T in enumerate(times[k:]):
            if np.isclose(t, T):
                dy = Y-H@U[k+_]
                J_obs += dy@Rinv@dy/2.0
                k += _+1
                break
    print(J_0, J_obs)
    return J_0 + J_obs
                
def grad_J_4d_var(u, u_b, dt, y, t_obs, t_0):
    """Calculate 4D-Var cost function gradient using continuous adjoint
    
    Parameters
    ----------
    
    u : arraylike
        test state
    u_b : arraylike
        background state
    dt : float
        timestep
    y : list
        Collection of observations
    t_obs: arraylike
        Validity times for the observations
    t_0 : float
        start time
    
    Returns
    -------
    nabla_J : arraylike
        4D-var cost function gradient
    """
    dJ_obs = continuous_adjoint_equation(u, dt, t_obs[-1], y.copy(), t_obs.copy(), t_0)
    return Binv@(u-u_b)+dJ_obs


from scipy.optimize import minimize
def continuous_4d_var(u_b, y, dt, t_obs, t_0=0.0):
    """Calculate 4D-Var analysis
    
    Parameters
    ----------
    u_b : arraylike
        background state
    y : list
        Collection of observations
    dt : float
        timestep
    t_obs: arraylike
        Validity times for the observations
    t_0 : float
        start time
    
    Returns
    -------
    y_a : arraylike
        Analysis state
    """
    u = u_b.copy()
    res = minimize(J_4d_var, u, jac=grad_J_4d_var, args=(u_b, dt, y, t_obs, t_0),
                   method='CG', tol = 1e-3, options={'maxiter':100})
    return res.x
    print(J_4d_var(u_b, u_b, dt, y, t_obs, t_0))
    u=u_b.copy()
    for _ in range(5):
        u = u - 0.005*grad_J_4d_var(u, u_b, dt, y, t_obs, t_0)
        print(J_4d_var(u, u_b, dt, y, t_obs, t_0))
    return grad_J_4d_var(u_b.copy(), u_b, dt, y, t_obs, t_0)

u0 = np.ones(nz)
# note that we deliberately reduce the timestep here to deal with the stability problem.
u_a = continuous_4d_var(u0, y, 0.1*dt, t_obs)
plt.plot(u_a)
```

Note that if we chose a more sensible discretization we could use a longer timestep. In particular, if we used the same approach as the forward model then for this system the discretized continous adjoint would be the same as the discrete adjoint.

#### Discontinuous adjoint 4D-var

For this method we take the (nonlinear) discrete model and linearize it directly. This can be more work, and gives more opportunities for mistakes, but the 

The forward model equation can be written in the form
$$ \mathbf{A} \mathbf{u}^{k+1} = \mathbf{l}(\mathbf{u}) + \mathbf{f} $$

where 

$$ A = \left(
\begin{array}{ccccc}
1+\frac{\Delta t\nu}{\Delta z^2}& -\frac{\Delta t\nu}{2\Delta z^2}& 0& \ldots &-\frac{\Delta t\nu}{2\Delta z^2} \\
-\frac{\Delta t\nu}{2\Delta z^2}&1+\frac{\Delta t\nu}{\Delta z^2}& -\frac{\Delta t\nu}{2\Delta z^2}&\ldots &0 \\
0&-\frac{\Delta t\nu}{2\Delta z^2}&1+\frac{\Delta t\nu}{\Delta z^2}& -\frac{\Delta t\nu}{2\Delta z^2},\ldots &0 \\
\vdots&&&&\vdots\\
-\frac{\Delta t\nu}{2\Delta z^2}&0&0&\ldots 1+\frac{\Delta t\nu}{\Delta z^2}
\end{array}
\right)
$$
and

$$l_i = u_i + \Delta t\left(\frac{u_{i-1}^2 -u_{i+1}^2}{4\delta z}+\nu\frac{u_{i+1}-2u_i+u_{i-1}}{2\Delta z^2}\right)$$

Alternatively

$$ \mathbf{u}^{k+1} = \mathbf{A}^{-1} \left(\mathbf{l}(\mathbf{u}) + \mathbf{f}\right) $$

As in the continuous case, we can linearize this using `sympy` and the identity 

$$ \mathbf{L}\delta \mathbf{u} = \left[\frac{d}{da} \mathbf{l}(\mathbf{u}+a\delta \mathbf{u})\right]_{a=0}. $$

```python
z = sympy.Symbol(r'z')
delz = sympy.Symbol(r'\Delta z')
delt = sympy.Symbol(r'\Delta t')
u_i = sympy.Function(r'u_i')(z,t)
u_p = sympy.Function(r'u_{i+1}')(z,t)
u_m = sympy.Function(r'u_{i-1}')(z,t)
du_i = sympy.Function(r'\delta u_i')(z,t)
du_p = sympy.Function(r'\delta u_{i+1}')(z,t)
du_m = sympy.Function(r'\delta u_{i-1}')(z,t)

#Burgers' equation

L_discrete = u_i + delt*((u_m**2-u_p**2)/(4*delz)+nu/2*(u_p-2*u+u_m)/delz**2)

# The tangent linear model
sympy.Equality(sympy.Derivative(L_discrete.subs({u_i:u_i+a*du_i,
                                                 u_p:u_p+a*du_p,
                                                 u_m:u_m+a*du_m}),
                                a, evaluate=True).subs(a,0), 0)
```

Hence

$$ L = \left(
\begin{array}{ccccc}
1-\frac{\Delta t\nu}{\Delta z^2}& -\frac{u_2}{2\Delta z}+\frac{\Delta t\nu}{2\Delta z^2}& 0& \ldots &\frac{u_N}{2\Delta z}+\frac{\Delta t\nu}{2\Delta z^2} \\
\frac{u_1}{2\Delta z}+\frac{\Delta t\nu}{2\Delta z^2}&1-\frac{\Delta t\nu}{\Delta z^2}& -\frac{u_3}{2\Delta z}+\frac{\Delta t\nu}{2\Delta z^2}&\ldots &0 \\
0&\frac{u_2}{2\Delta z}+\frac{\Delta t\nu}{2\Delta z^2}&1-\frac{\Delta t\nu}{\Delta z^2}& -\frac{u_4}{2\Delta z}+\frac{\Delta t\nu}{2\Delta z^2},\ldots &0 \\
\vdots&&&&\vdots\\
-\frac{u_1}{2\Delta z}-\frac{\Delta t\nu}{2\Delta z^2}&0&0&\ldots 1+\frac{\Delta t\nu}{\Delta z^2}
\end{array}
\right)$$


Again no forcing term appears.

The adjoint model is derived by setting

$$ \mathcal{J}_{ext} := \mathcal{J}+\sum_k (\mathbf{\lambda}^{(k)})^T (A\delta \mathbf{u}^{(k+1)}- L\delta \mathbf{u}^{(k)})$$

differentiating with respect to $\delta \mathbf{u}^{(k)}$ gives us 

$$ \mathbf{A}^T\mathbf{\lambda} ^{(k)} = L^T \mathbf{\lambda} ^{(k+1)} $$

if there are no observations on that time level and

$$ \mathbf{A}^T\mathbf{\lambda} ^{(k-1)} = L^T \mathbf{\lambda} ^{(k)} + \mathbf{H}^T \mathbf{R}^{-1}(\mathbf{y}^{(k)}-H\mathbf{u}^{(k)}) $$
if there are.

Note that we can get an alternative, but functionally equivalent model by using multiplier terms of the form
$$(\mathbf{\lambda}^{(k)})^T (\delta \mathbf{u}^{(k+1)}- A^{-1}L\delta \mathbf{u}^{(k)}) $$
using this form will modify whether the $A^{-T}$ happens at the start or end of the update, though not the final result. 


In this case we will code up the entire operation $\mathbf{A}^{-T}\mathbf{L}^T\mathbf{\lambda}$ as a function directly.

```python
def lhs_adj_matrix(nz, dt):
    """ Calculate A matrix.
    
    Parameters
    ----------
    nz : float
        Matrix size
    dt: float
        timestep
        
    Returns
    -------
    A : 2D ndarray
        Matrix for lhs of discretization
    """
    mat = np.eye(nz)
    mat[0,-1] -= 0.5*dt*nu/dz**2
    mat[0, 0] += dt*nu/dz**2
    mat[0, 1] -= 0.5*dt*nu/dz**2
    for _ in range(1, nz-1):
        mat[_, _-1] -= 0.5*dt*nu/dz**2
        mat[_, _] += dt*nu/dz**2
        mat[_, _+1] -= 0.5*dt*nu/dz**2
    mat[-1,-2] -= 0.5*dt*nu/dz**2
    mat[-1, -1] += dt*nu/dz**2
    mat[-1, 0] -= 0.5*dt*nu/dz**2
    return mat

amat = lhs_adj_matrix(nz, dt)
amatTinv = inv(amat.T)

def adjoint_step_solve(u, l, dt, obs_term=0):
    """ Calculate adjoint update for one timestep.
    
    Parameters
    ----------
    u : arraylike
        Burgers state
    l : arraylike
        current perturbation value
    dt : float
        timestep
    obs_term : arraylike
        Optional increment from difference to observations.
        
    Returns
    -------
    l_new : 2D ndarray
        New perturbation value.
    """
    L_T_lam = np.empty(l.shape)
    ### note we could switch the sign of u to -u and the order of l[:2]-l[2:] to l[2:]-l[:2]
    L_T_lam[0] = l[0]+dt*(u[0]*(l[1]-l[-1])/(2*dz)
                                +0.5*nu*(l[-1]-2*l[0]+l[1])/dz**2)
    L_T_lam[1:-1] = l[1:-1]+dt*(u[1:-1]*(l[2:]-l[:-2])/(2*dz)
                                +0.5*nu*(l[:-2]-2*l[1:-1]+l[2:])/dz**2)
    L_T_lam[-1] = l[-1]+dt*(u[-1]*(l[0]-l[-2])/(2*dz)
                                +0.5*nu*(l[-2]-2*l[-1]+l[0])/dz**2)
    
    return amatTinv.dot(L_T_lam+obs_term)

def discontinuous_adjoint_equation(u0, dt, t_final, y, t_obs, t_0=0.0):
    """ Solve the discontinuous adjoint equation for a 4d var model.
    
    Parameters
    ----------
    u0 : arraylike
        The starting condition
    dt :  float
        The timestep
    t_final: float
        The final time
    y : list
        a collection of observations
    t_obs: list
        the validity times of the observations
    t_0 : float
        the start time
        
    Returns
    -------
    l :
    The update valid at t_0
    """
    u, times = burgers_equation(u0, dt, t_final, t_0)
    l = 0*u[-1]
    mat = lhs_matrix(l, dt)
    matinv = inv(mat)        
    
    for U, t in zip(u[::-1], times[::--1]):
            if np.isclose(t, t_obs[-1]):
                l = adjoint_step_solve(U, l, dt, -H.T@Rinv@(y[-1]-H@U))
                t_obs = t_obs[:-1]
                y = y[:-1]
            else:
                l = adjoint_step_solve(U, l, dt)
    return l

## Now define the standard 4d-var cost function and its gradient
z = np.linspace(-10, 10, nz)
B = generate_B_matrix(z)
Binv = inv(B)
Rinv = inv(R)

def J_4d_var(u, u_b, dt, y, t_obs, t_0):
    """Calculate 4D-Var cost function
    
    Parameters
    ----------
    
    u : arraylike
        test state
    u_b : arraylike
        background state
    dt : float
        timestep
    y : list
        Collection of observations
    t_obs: arraylike
        Validity times for the observations
    t_0 : float
        start time
    
    Returns
    -------
    J : float
        4D-var cost function value
    
    """
    U, times = burgers_equation(u, dt, t_obs[-1], t_0)
    J_0 = (u-u_b)@Binv@(u-u_b)/2.0
    J_obs = 0.0
    k = 0
    for Y, t in zip (y, t_obs):
        for _, T in enumerate(times[k:]):
            if np.isclose(t, T):
                dy = Y-H@U[k+_]
                J_obs += dy@Rinv@dy/2.0
                k += _+1
                break
    print(J_0, J_obs)
    return J_0 + J_obs
                
def grad_J_4d_var(u, u_b, dt, y, t_obs, t_0):
    """Calculate 4D-Var cost function gradient using discontinuous adjoint
    
    Parameters
    ----------
    
    u : arraylike
        test state
    u_b : arraylike
        background state
    dt : float
        timestep
    y : list
        Collection of observations
    t_obs: arraylike
        Validity times for the observations
    t_0 : float
        start time
    
    Returns
    -------
    nabla_J : arraylike
        4D-var cost function gradient
    """
    
    dJ_obs = discontinuous_adjoint_equation(u, dt, t_obs[-1], y, t_obs, t_0)
    return Binv@(u-u_b)+dJ_obs


from scipy.optimize import minimize
def discontinuous_4d_var(u_b, y, dt, t_obs, t_0=0.0):
    """Calculate 4D-Var analysis
    
    Parameters
    ----------
    u_b : arraylike
        background state
    y : list
        Collection of observations
    dt : float
        timestep
    t_obs: arraylike
        Validity times for the observations
    t_0 : float
        start time
    
    Returns
    -------
    y_a : arraylike
        Analysis state
    """
    res = minimize(J_4d_var, u_b.copy(), jac=grad_J_4d_var, args=(u_b, dt, y, t_obs, t_0),
                   method='CG', tol = 1e-3, options={'maxiter':100})
    return res.x

u0 = np.ones(nz)
u_a = discontinuous_4d_var(u0, y, dt, t_obs)
plt.plot(u_a)
    
```

We have used a single window approach in both these cases. For a multiple window approach we would just need to split the observations into bins, and then use the final forecast value of the previous window as the background state of the new window.

#### Multiple OI assimilations

At each time level where we have observations we will use the OI update formula

$$\mathbf{x}_a = \mathbf{x}_b + \mathbf{B} \mathbf{H}^T\left( \mathbf{R} + \mathbf{H}\mathbf{B}\mathbf{H}^T\right)^{-1} (\mathbf{y}-\mathbf{H}\mathbf{x}_b).$$

Note that this is easier to use in this situation than it might be in others since $\mathbf{B}$, $\mathbf{R}$ and $\mathbf{H}$ are all fixed for every observation timestep.

```python
# In the OI worldview, the B matrix is fixed, so we can just calculate this once
z = np.linspace(-10, 10, nz)
B = generate_B_matrix(z)
W = B@H.T@inv(R + H@B@H.T)

def OI_assimilation_step(x_b, y, *args):
    """ Do single OI update.
    
    Parameters
    ----------
    x_b : arraylike
        background state
    y : arraylike
        observations at timelevel

    Returns
    x_a : arraylike
        analysis state
    
    """
    return x_b + W@(y-H@x_b)

def OI(u0, y, t_obs, t_final, t_0=0.0):
    """ Do interated OI updates.
    
    Parameters
    ----------
    u0 : arraylike
        start state
    y : list
        Collection of observations
    t_obs : arraylike
        Validity time of observations
    t_final : float
        finish time
    t_0 : float
        start time

    Returns
    x_a : arraylike
        analysis state
    
    """
    u, times = burgers_equation(u0, dt, t_obs[0], t_0)
    for _ in range(len(y)-1):
        u_a = OI_assimilation_step(u[-1], y[_])
        u, times = burgers_equation(u_a, dt, t_obs[_], times[-1])
    u_a = OI_assimilation_step(u[-1], y[-1])
    u, times = burgers_equation(u_a, dt, t_final, times[-1])
    return u[-1]

u0 = np.ones(nz)
u_a = OI(u0, y, t_obs, dt)
plt.plot(u_a)
```

#### Multiple 3D-Var assimilations 

Since we have arranged matters for the observation operator to be linear in this case, the method is essentially identical to the previous case. However we can now either express the problem in terms of a quadratic cost function and use gradient based minimization to find the solution, or just write down the solution to $\nabla \mathcal{J}=0$. Due to some relatively large matrix vector mupliplication, the OI approach is likely to prove faster in this case.

```python
z = np.linspace(-10, 10, nz)
B = generate_B_matrix(z)
Binv = inv(B)
Rinv = inv(R)

def J(u, u_b, y):
    """Calculate 3D Var cost function
    
    Parameters
    ----------
    u : arraylike
        test state
    u_b : arraylike
        background state
    y : arraylike
        vector of observations
    
    Returns
    -------
    J : float
        Value of cost function 
    """
    dx_b = u-u_b
    dx_o = y - H@u
    return 0.5*(np.dot(dx_b, Binv.dot(dx_b))+np.dot(dx_o,Rinv.dot(dx_o)))

def grad_J(u, u_b, y):
    """Calculate 3D Var cost function gradient
    
    Parameters
    ----------
    u : arraylike
        test state
    u_b : arraylike
        background state
    y : arraylike
        vector of observations
    
    Returns
    -------
    nabla_J : arraylike
        Gradient of cost function 
    """
    dx_b = u-u_b
    dx_o = y - H@u
    return Binv@dx_b-H.T@Rinv@dx_o

from scipy.optimize import minimize

def do_3D_var_step(u_b, y, *args):
    """Calculate single 3D Var update.
    
        Parameters
    ----------
    u_b : arraylike
        background state
    y : arraylike
        observations at timelevel

    Returns
    x_a : arraylike
        analysis state
    """
    res = minimize(J, u_b.copy(), jac=grad_J, method='CG',
                   args=(u_b,y), tol = 1e-3, options={'maxiter':100})
    return res.x

def do_3D_var(u0, y, t_obs, t_final, t_0=0.0):
    """ Do interated OI updates.
    
    Parameters
    ----------
    u0 : arraylike
        start state
    y : list
        Collection of observations
    t_obs : arraylike
        Validity time of observations
    t_final : float
        finish time
    t_0 : float
        start time

    Returns
    x_a : arraylike
        analysis state
    
    """
    u, times= burgers_equation(u0, dt, t_obs[0], t_0)
    for _ in range(len(y)-1):
        u_a = do_3D_var_step(u[-1], y[_])
        u, times = burgers_equation(u_a, dt, t_obs[_], times[-1])
    u_a = do_3D_var_step(u[-1], y[-1])
    u, times = burgers_equation(u_a, dt, t_final, times[-1])
    return u[-1]

u0 = np.ones(nz)
u_a = do_3D_var(u0, y, t_obs, dt)
plt.plot(u_a)
```

You will note that this is the same as the OI method above.

### Driver code

we can descrive some fairly generic code for the generic combined forecast/assimilation system. The general pattern is:
1. assimilate the available data,
2. forecast until we have new data to assimilate,
3. repeat if necesary.

For the single window 4D-Var, the outer loop only happens once. For multiple window 4D-var, it happens for the few windows, for the other two methods, it happens every time we have new observations.
